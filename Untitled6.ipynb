```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression

# Load the dataset
data = pd.read_csv('Shopping_Trends.csv')

# Step 1: Exploratory Data Analysis (EDA)
print("Dataset Overview:\n", data.head(), "\n")
print("Dataset Shape:", data.shape)
print("Dataset Description:\n", data.describe())
print("Missing Values:\n", data.isnull().sum())

# Visualize missing data
plt.figure(figsize=(10, 5))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()

# Step 2: Data Cleaning and Preprocessing
# Fill missing values for numerical and categorical features
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

data[numerical_cols] = data[numerical_cols].fillna(data[numerical_cols].median())
data[categorical_cols] = data[categorical_cols].fillna("Unknown")

# Encode categorical variables
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le  # Save encoders for reverse transformation if needed

# Normalize numerical features
scaler = MinMaxScaler()  # Using Min-Max Scaling
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Feature Engineering: Create additional features
data['High-Spender'] = (data['Purchase Amount (USD)'] > data['Purchase Amount (USD)'].mean()).astype(int)
data['Discounted'] = (data['Review Rating'] > 4).astype(int)  # Example feature

# Correlation Analysis
plt.figure(figsize=(15, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Step 3: Feature Selection and Dimensionality Reduction
# Separate features and target
X = data.drop(['Purchase Amount (USD)', 'High-Spender'], axis=1)  # Exclude target and engineered features
y = data['Purchase Amount (USD)']

# Select top-k features
select_k_best = SelectKBest(score_func=f_regression, k=10)
X_selected = select_k_best.fit_transform(X, y)

# Dimensionality reduction using PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_selected)
print("Explained Variance Ratio by PCA Components:", pca.explained_variance_ratio_)

# Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Step 5: Model Training and Hyperparameter Tuning
# Define the model
model = RandomForestRegressor(random_state=42)

# Hyperparameter tuning using GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='r2', verbose=2)
grid_search.fit(X_train, y_train)


```
